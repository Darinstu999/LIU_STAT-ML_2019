```python
# remove the puncs and split it into sep items
def get_text(txtFile):
    f = open(txtFile)
    text = f.read().lower()
    for i in '!.?,*':
        text = text.replace(i,' ')
    return text.split()

import numpy as np

# when the input is raw=get_text(txtFile) and startWord is defined by user
# it returns a list of the Word B's freq followed Word A
def follow(raw,startWord):
    temp = np.array(raw)
    indexTuple = np.where(temp == startWord)
    indexList = list(indexTuple)
    index1 = [i + 1 for i in indexList]
    out = []
    for i in index1:
        out.append(temp[i].tolist())
    out1 = out[0]
    
    counts = {}

    for i in out1:
        counts[i] = counts.get(i,0) + 1

    iteams = list(counts.items())
    return iteams

# sideway func
def takeSecond(elem):
    return elem[1]
# from the list above, select one Word as WordB
def getB(a):
    a.sort(key = takeSecond, reverse = True)
    return a[1][0]
        
# set length as input to control the length of output, and give B to A after every loop
# generate a list, then use .join to conbine them
def gen(txtFile, startWord, length):
    
    raw = get_text(txtFile)
    
    out2 = [startWord]
    
    while len(out2) <= length:
        temp = follow(raw,startWord)
        wordB = getB(temp)
        out2.append(wordB)
        startWord = wordB
    out2 = ' '.join(out2)
    return out2

gen(txtFile='/Users/darin/Desktop/python/shakespeare.txt',
   startWord='king',
   length = 30)
```

```python
import sys
import re
import os.path
from collections import Counter
```

```python
def get_text(file):
    text = file.read().lower()
    for i in '!.?,*':
        text = text.replace(i, ' ')
    return text.split()
```

```python
a = get_text(txtFile='/Users/darin/Desktop/python/shakespeare.txt')
```

```python
def output(line):
    print(line)
    if len(sys.argv[1:]) == 2:
        with open(sys.argv[2], "a", encoding="utf8") as output:
            output.write(line + "\n")
```

```python
def letter_frequency(doc):
    c = Counter(doc)
    title = "Frequency table for alphabetic letters:"
    output(title)
    for letter in c.most_common():
        content = f"{letter[0]} ({letter[1]} occurrences)"
        output(content)
```

```python
letter_frequency(a)
```

```python
def number_of_words(doc):
    c = Counter(doc)
    content = f"The total number of words is {sum(c.values())}"
    output(content)
```

```python
number_of_words(a)
```

```python
def number_of_unique_words(doc):
    c = Counter(doc)
    content = f"The number of unique words is {len(c)}"
    output(content)
```

```python
number_of_unique_words(a)
```

```python
def common_words(doc):
    c = Counter(doc)
    five_most_common = dict(c.most_common(5))
    title = "The five most common words are: "
    output(title)
    for key, value in five_most_common.items():
        three_most_common = dict(c.most_common(3))
        content = f"{key} ({value} occurrences)"
        output(content)
        for key2, value2 in three_most_common.items():
            content2 = f"-- {key2}, {value2}"
            output(content2)
```

```python
common_words(a)
```

```python
def gen(doc):
    cur_word = sys.argv[2]
    msg = cur_word
    max_number = int(sys.argv[3])
    number_of_words = 1

    c = dict(Counter(doc))

    if cur_word in doc:
        while number_of_words < max_number and cur_word != doc[-1]:
            c2 = Counter(doc)
            successors = dict(c2.most_common())
            for key, value in successors.items():
                successors = {key: value / c[cur_word]}
            selected_successor = choices(population=list(successors.keys()),
                                         weights=list(successors.values()), k=1)
            cur_word = selected_successor[0]
            msg = msg + " " + cur_word
            number_of_words += 1

    return msg
```

```python
gen(a)
```
